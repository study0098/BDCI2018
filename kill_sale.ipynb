## replace,agg
- 对df中的表格寻找 某个形式的字符串（并进行替换）时，波兰表达式不大好弄
- 所以df.replace不好直接用
- 所以写一个小函数，对传入的字符串进行replace(',','')，然后把这个小函数应用到df的所有元素上，返回一个shape不变的df
- 同理，agg(np.float)

```python
def ff(st):
    return st.replace(',','')
```

## 算出同一个good内部，差值、比例 与 sku销量 的相关性

```python
def quan(df):
    return df.quantile()
```

```python
dff = DataFrame()
for i in range(0,num):
    goods_id = sale_head.iloc[i,0]
    print(goods_id)
    tt = sale_o.loc[sale_o['goods_id']==goods_id,:]
    tt0 = tt.groupby(tt['sku_id'],as_index=False)['比例'].mean()
    tt0.columns = ['sku_id','比例_mean']
    print(tt0)
    tt1 = tt.groupby(tt['sku_id'],as_index=False)['比例'].median()
    tt1.columns = ['sku_id','比例_quan']
    print(tt1)
    tt_01 = pd.merge(tt0,tt1,on='sku_id')
    sale_good = goods_sku_sale_total.loc[goods_sku_sale_total['goods_id']==goods_id,:]
    mm = pd.merge(sale_good,tt_01,on='sku_id')
    if i==0:
        dff = mm
    else:
        dff = pd.concat([dff,mm])
dff = dff.reset_index().drop('index',axis=1)
```

```python
dff
```

```python
def mycorr(df,col_a,col_bs):
    ll = []
    for i in col_bs:
        a = df[col_a].corr(df[i])
        ll.append(a)
    return Series(ll,index=col_bs)
```

```python
re = dff.groupby(dff.goods_id).apply(mycorr,col_a='goods_num',col_bs=['比例_mean','比例_quan'])
```
