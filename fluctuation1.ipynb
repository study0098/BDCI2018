```python
import pandas as pd
from tqdm import tqdm
import numpy as np
goodsSale = pd.read_csv('C:/Users/xyttttt/Desktop/alldata/goodsale.csv',encoding='utf8',low_memory=False)
submit = pd.read_csv('C:/Users/xyttttt/Desktop/alldata/submit_example.csv',encoding='utf8',low_memory=False)

goodsNum = pd.DataFrame({'sku_id':submit['sku_id']})
goodsNum12 = pd.DataFrame({'sku_id':submit['sku_id']})
dates = []
for i in range(20171214,20171232):
    dates.append(i)
for date in dates:
    sale12 = goodsSale[(goodsSale.data_date==date)]
    df = pd.DataFrame({str(date):sale12.goods_num,'sku_id':sale12.sku_id})
    goodsNum12 = pd.merge(goodsNum12,df,on='sku_id',how='left')
plandate = []
for date in range(20171214,20171223):
    plandate.append(str(date))
goodsNum12['3'] = pd.DataFrame(goodsNum12,columns=plandate).apply(lambda x:x.sum(),axis=1)
plandate = []
for date in range(20171223,20171227):
    plandate.append(str(date))
goodsNum12['4'] = pd.DataFrame(goodsNum12,columns=plandate).apply(lambda x:x.sum(),axis=1)
plandate = []
for date in range(20171227,20171232):
    plandate.append(str(date))
goodsNum12['5'] = pd.DataFrame(goodsNum12,columns=plandate).apply(lambda x:x.sum(),axis=1)

goodsNum12
```
