```python
import pandas as pd
from tqdm import tqdm
import numpy as np
goodsSale = pd.read_csv('C:/Users/xyttttt/Desktop/alldata/goodsale.csv',encoding='utf8',low_memory=False)
submit = pd.read_csv('C:/Users/xyttttt/Desktop/alldata/submit_example.csv',encoding='utf8',low_memory=False)

goodsNum = pd.DataFrame({'sku_id':submit['sku_id']})
goodsNum12 = pd.DataFrame({'sku_id':submit['sku_id']})
dates = []
date1=[]
for i in range(20170522,20170532):
    dates.append(i)
    date1.append(str(i))
for i in range(20170601,20170627):
    dates.append(i)
    date1.append(str(i))
for date in dates:
    sale12 = goodsSale[(goodsSale.data_date==date)]
    df = pd.DataFrame({str(date):sale12.goods_num,'sku_id':sale12.sku_id})
    goodsNum12 = pd.merge(goodsNum12,df,on='sku_id',how='left')
goodsNum12=goodsNum12.fillna(0.0)
goodsNum12['total']=pd.DataFrame(goodsNum12,columns=date1).apply(lambda x:x.sum(),axis=1)
goodsNum12
```
